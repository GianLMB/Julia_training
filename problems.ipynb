```julia
# Problem 3: Sierpinski's triangle

function sierpinski(n)  
    px=[0,1,0]
    py=[0,0,1]
    function rec(n,px,py)
        n == 1 && return px,py 
        a =  2^(n-2)
        psub = rec(n-1,px,py)
        pright = rec(n-1,px+a*ones(length(px)),py)
        pup = rec(n-1,px,py+a*ones(length(py)))
        px = [psub[1]; NaN; pright[1]; NaN; pup[1]]
        py = [psub[2]; NaN; pright[2]; NaN; pup[2]]
        return px,py
    end
    rec(n,px,py)
end

px,py = sierpinski(9)

using Plots
gr()
plot(px,py,seriestype=:shape,size=(600,600),linewidth=-1,grid=false,axis=false,label=false)
```

```julia
# Problem 7: In-place quicksort

function partition!(vec,i,j)   # select a pivot p = vec[i] and divide vec in 3: < p, p , >= p; returns p final index
    pval = vec[i]
    pindex = i
    for k = i+1:j
        if vec[k] < pval
            aux = vec[pindex+1]
            vec[pindex] = vec[k]
            vec[pindex+1] = pval
            vec[k] = aux
            pindex += 1
        end
    end
    return pindex
end 
        
function qsort!(vec,i,j)       # ! in julia is used for in-place functions 
    j-i <= 1 && return 
    pivot = partition!(vec,i,j)       
    qsort!(vec,i,pivot-1)
    qsort!(vec,pivot+1,j)
end

v = [1,2,3,4,-5,7,41,-3]
qsort!(v,1,length(v))
v
```

```julia
# Problem 8: argmax for maximum subarray
# (modifying Example 4)

function maxsub(v)
    n = length(v)
    L = v[1]
    M = L
    I,J = 1,1
    for j = 2:n
        L = max(L+v[j], v[j])        
        M = max(L,M)
        if L == v[j]
            I = j
        end
        if M == L
            J = j
        end
    end
    return M,I,J
end
    
v = [1,2,3,-5,-2]
i,j = maxsub(v)    
```

```julia
# Problem 9: 2D maximum subarray

function sumcol(mat)
    n, m = size(mat)
    sumvec = zeros(n)
    for i = 1:n
        for j = 1:m
            sumvec[i] += mat[i,j]
        end
    end
    return sumvec
end 

function D2maxsub(mat)
    n , m = size(mat)
    k = zeros(Int128,(m,m))
    l = zeros(Int128,(m,m))
    maxsum = zeros(Int128,(m,m))
    for q = 1:m
        for p = 1:q
            colsum = sumcol(mat[:,p:q])
            maxsum[p,q], k[p,q], l[p,q] = maxsub(colsum)
        end
    end
    maximum, pmax, qmax = max(maxsum)
    kmax = k[pmax,qmax]
    lmax= l[pmax,qmax]
    return maximum, kmax, lmax, pmax, qmax
end
            
```

```julia
# Problem 10: Divide and Conquer max_subarray

function div_maxsub(a)
    n = length(a)
    n == 0 && return
    n == 1 && return a[1],a[1],a[1],a[1]
    m = div(n,2)
    al = @view a[1:m]        # similar to copy(), but uses less memory
    ar = @view a[m+1:end] 
    M_ac,L_ac,R_ac,S_ac = div_maxsub(al) 
    M_cb,L_cb,R_cb,S_cb = div_maxsub(ar) 
    S = S_ac+S_cb
    L_ab = max(L_ac,L_cb+S_ac)
    R_ab = max(R_cb,R_ac+S_cb)
    M_ab = max(M_ac,M_cb,R_ac+L_cb)
    println(M_ab)
    return M_ab,L_ab,R_ab,S
end

a = [1,4,-15,2,3,7,5,2]
div_maxsub(a)
```

```julia
# Problem 11: Recursive Subsets

function subsets(n,k)
    
    if k == n
        return [1:n;]'
    end
    if k == 1
        return [1:n;]
    end
    
    aset = subsets(n-1,k)
    bset = subsets(n-1,k-1)
    
    if size(bset,1) == 0
        set = aset
    else
        set = vcat(aset,hcat(bset,ones(Int,size(bset,1))*n))
    end
    return set
end  

subsets(5,3)
```

```julia
# Problem 12: growing subsequences

function subl(v)
    n = length(v)
    function L(i,v)
        i == 1 && return 1
        bool = true
        for j = 1:i
           if v[j] < v[i]
              bool = false
           end
        end
        bool == true && return 1
        return 1 + maximum(L(j,v) for j = 1:i-1 if v[j] <= v[i])
    end
    maximum(L(i,v) for i = 1:n)
end

v = [1, 4, -1, 3, 6, 2, -3, 10, -2, 11, 1, 2, 4, 7, 10]
subl(v)
```
